```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data_Mazancieux_2018.csv')
```

```python
df.head()
```

### Line Plot - Confidence Trayectories per task - Only 1 participant

```python
participant = 1

df_participant = df[df['Subj_idx'] == participant].copy()
df_participant.loc[:, 'Confidence_standardized'] = df_participant.groupby('Task')['Confidence'].transform(lambda x: (x - x.mean()) / x.std())

plt.figure(figsize=(10,6))
for task, group_data in df_participant.groupby('Task'):
    x_values = range(1, len(group_data) +1)
    plt.plot(x_values, group_data['Confidence_standardized'], marker='o', label=task)

plt.xlabel('Trials')
plt.title(f'Participant {participant}')
plt.ylabel('Confidence_standardized')
plt.legend(title = 'Task')
plt.show()
```
